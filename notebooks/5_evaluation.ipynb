```python
# notebooks/5_evaluation.ipynb

# Gerekli kütüphaneyi yükleyelim (Eğer yüklü değilse)
# !pip install shap

import shap
import xgboost as xgb
import pandas as pd
import matplotlib.pyplot as plt

# 1. Veriyi Oku
df = pd.read_csv('../data/atm_processed.csv')

# 2. Modeli Yükle
model = xgb.XGBRegressor()
model.load_model('../models/atm_optimized_model.json')

# 3. Features Seçimi (KRİTİK ADIM)
# Model eğitilirken hangi sırayı kullandıysak AYNISINI kullanmalıyız.
features = ['hour', 'day_of_week', 'day_of_month', 'is_weekend', 
            'lag_1', 'lag_24', 'rolling_mean_3', 'rolling_mean_24']

X = df[features]

# 4. SHAP Explainer Oluştur
# Veri seti büyük olduğu için sadece ilk 1000 satırla analiz yapıyoruz (Hız için yeterli)
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X.iloc[:1000])

# 5. Grafiği Çiz ve KAYDET
plt.figure(figsize=(10, 6))
plt.title("Hangi Özellik Tahmini Nasıl Etkiliyor? (SHAP Summary)")

# show=False diyoruz ki hemen ekrana basıp kapatmasın, üzerinde işlem yapalım
shap.summary_plot(shap_values, X.iloc[:1000], show=False)

# Grafiği dosya olarak kaydet (App.py'da kullanacağız)
plt.savefig('shap_summary.png', bbox_inches='tight', dpi=300)
print("✅ Grafik başarıyla kaydedildi: notebooks/shap_summary.png")

# Şimdi ekranda göster
plt.show()
```
